```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Agriculture', 'Energy', 'Industry', 'Transport', 'Others']
sizes = [30, 25, 20, 15, 10]  # Percentages of the pie chart
# This data is not from any source its an assupmtion based on article read on greenhouse gas sectors
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']  # Colors for each section
explode = (0.1, 0, 0, 0, 0)  # Explode the first slice (Agriculture)

# Create a pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

# Ensure pie is drawn as a circle
plt.axis('equal')

# Add a title
plt.title("Greenhouse Gas Emissions by Sector")

# Show the pie chart
plt.show()
```
